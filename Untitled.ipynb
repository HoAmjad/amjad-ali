```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('hotel.csv')
```

```python
print(df.columns)
```

```python
plt.plot(df['adults'], df['children'])  
plt.xlabel('Data')
plt.ylabel(' Bookings by ages')
plt.title('Line Graph: adult vs. children')
plt.show()
```

```python
plt.plot(df['lead_time'], df['adr'], color='blue', linestyle='-', linewidth=1)
plt.xlabel('Lead Time (Days)')
plt.ylabel('Average Daily Rate (ADR)')
plt.title('Lead Time vs Average Daily Rate')
plt.show()
```

```python
plt.plot(
    df['lead_time'], df['adr'], 
    color='green', linestyle='--', linewidth=2, 
    marker='o', markerfacecolor='blue', markersize=8
)
plt.xlabel('Lead Time (Days)')
plt.ylabel('Average Daily Rate (ADR)')
plt.title('Customized Line Plot')
plt.ylim(0, df['adr'].max())  # Adjust Y-axis range
plt.xlim(0, df['lead_time'].max())  # Adjust X-axis range
plt.show()
```

```python
df_grouped = df.groupby('market_segment')['adr'].mean()
plt.bar(df_grouped.index, df_grouped.values, color='orange')
plt.xlabel('Market Segment')
plt.ylabel('Average Daily Rate (ADR)')
plt.title('Vertical Bar Chart: Market Segment vs ADR')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.barh(df_grouped.index, df_grouped.values, color='lightblue')
plt.xlabel('Average Daily Rate (ADR)')
plt.ylabel('Market Segment')
plt.title('Horizontal Bar Chart: Market Segment vs ADR')
plt.show()
```

```python
plt.hist(df['lead_time'], bins=10, color='purple', alpha=0.7)
plt.xlabel('Lead Time (Days)')
plt.ylabel('Frequency')
plt.title('Histogram of Lead Time')
plt.show()
```

```python
plt.scatter(df['lead_time'], df['adr'], color='blue', marker='o')
plt.xlabel('Lead Time (Days)')
plt.ylabel('Average Daily Rate (ADR)')
plt.title('Scatter Plot: Lead Time vs ADR')
plt.show()
```

```python
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)
plt.plot(x, y, label='Sine Wave', color='cyan')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('Sine Curve')
plt.legend()
plt.show()
```
